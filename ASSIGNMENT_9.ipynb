###***1. What is Simple Linear Regression (SLR)? Explain its purpose .***

**Simple Linear Regression (SLR)** is a statistical method used to model the relationship between **one independent variable (X)** and **one dependent variable (Y)** by fitting a straight line to the observed data.

####***Definition -***

Simple Linear Regression assumes a linear relationship of the form :

                          Y = Œ≤‚ÇÄ + Œ≤‚ÇÅX + Œµ

Where :

*  **Y** = dependent (response) variable

* **X** = independent (predictor) variable

* **Œ≤‚ÇÄ (intercept)** = value of Y when X = 0

* **Œ≤‚ÇÅ (slope)** = change in Y for a one-unit change in X


* **Œµ** = error term (random noise)


###***Purpose of Simple Linear Regression***

####**(I) Understand Relationships -**

It helps determine **how strongly and in what direction X** affects Y

* Positive slope ‚Üí Y increases as X increases

*  Negative slope ‚Üí Y decreases as X increases


####**(II) Prediction -**

Once the model is fitted, it can be used to **predict the value of Y** for a given value of X

* Example : Predicting salary based on years of experience


####**(III) Quantify Impact -**

The slope (Œ≤‚ÇÅ) quantifies the **magnitude of change** in Y due to X

* Example : ‚ÄúFor every additional year of experience, salary increases by ‚Çπ50,000‚Äù


####**(IV) Trend Analysis -**

Identifies and summarizes trends in data using a simple, interpretable model

###***2. What are the key assumptions of Simple Linear Regression ?***

Simple Linear Regression (SLR) relies on a set of assumptions to ensure that the model is valid and that statistical inferences (like confidence intervals and hypothesis tests) are reliable. The key assumptions are :

####**(I) Linearity -**

* The relationship between the independent variable X and the dependent variable Y is linear.

* Mathematically :

             Y =  Œ≤0‚Äã + Œ≤1‚ÄãX + Œµ

* **Check :** Scatter plot of X vs Y.


####**(II) Independence of Errors -**

*  The residuals (errors) are **independent** of each other.

* Violations often occur in time-series data (autocorrelation).

*  **Check :** Durbin‚ÄìWatson test, residuals vs. time plot.


####**(III) Homoscedasticity (Constant Variance) -**

* The variance of the errors is **constant** across all values of **X**.

* If variance changes with **X**, it is called **heteroscedasticity**.

* **Check :** Residuals vs. fitted values plot.

####**(IV) Normality of Errors -**

* The residuals are **normally distributed** (mainly required for hypothesis testing and confidence intervals) .

* **Check :** Histogram of residuals, Q‚ÄìQ plot.


####**(V) Zero Mean of Errors -**

* The expected value of the error term is zero :

                      E(Œµ)=0         

*  Ensures the regression line is unbiased.



####**(VI) No Perfect Multicollinearity (Trivial in SLR) -**

* In SLR, there is only one predictor, so this assumption is automatically satisfied.

* Mentioned mainly for consistency with multiple regression.

###***5. What is the method of least squares in linear regression?***

**The Method of Least Squares** is a technique used in linear regression to estimate the model parameters (coefficients) so that the regression line best fits the observed data.



####**Core Idea -**

It finds the line (or hyperplane) that **minimizes the sum of the squared differences** between the observed values and the values predicted by the regression model .

These differences are called **residuals** .

###***8. What is the purpose of the R-squared metric in regression analysis ?***

**R-squared** (ùëÖ^2) measures how well a regression model explains the variability of the dependent (target) variable.

####**Purpose of R-squared**

*  It indicates the **proportion of variance in the dependent variable that is explained by the independent variables** in the model.

* It helps assess the **goodness of fit** of a regression model.



####**Interpretation**

* R^2 = 1 : The model explains **100%** of the variance (perfect fit) .

* R^2 = 0 : The model explains **none** of the variance .

* Example: R^2 = 0.80 means **80% of the variation** in the target variable is explained by the model.

####**Key Points**

* Higher R^2 generally indicates a better fit, but it **does not guarantee** the model is appropriate.

* **R^2 always increases** (or stays the same) when more predictors are added, even if they are not useful.

* For multiple predictors, **Adjusted R-squared** is preferred as it penalizes unnecessary variables.

###***9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.***
###***(Include your Python code and output in the code box below.)***

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y = np.array([15, 25, 35, 45, 55])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

This output means the fitted regression equation is :

###***10. How do you interpret the coefficients in a simple linear regression model ?***

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Independent variable
y = np.array([3, 5, 7, 9, 11])               # Dependent variable

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Extract coefficients
slope = model.coef_[0]
intercept = model.intercept_

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)

# Interpretation
print(f"\nInterpretation:")
print(f"For every 1-unit increase in X, Y increases by {slope}.")
print(f"When X = 0, the predicted value of Y is {intercept}.")
```
