**Question 1** : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

Answer:   Artificial Intelligence (AI) is a broad field focused on creating machines that can perform tasks that usually require human intelligence, such as decision-making, problem-solving, and understanding language. AI includes rule-based systems, robotics, computer vision, etc.

*   Machine Learning (ML) is a subset of AI. Instead of giving rules, ML allows systems to learn patterns from data. ML models improve automatically with experience. Examples include spam detection, price prediction, and recommendation systems.
*   Deep Learning (DL) is a specialized branch of ML that uses artificial neural networks with many layers. DL works extremely well with large datasets, image recognition, speech processing, and NLP.


*  Data Science is the field that works end-to-end with data — cleaning, analysis, visualization, modeling, and interpretation. It uses statistics, ML techniques, and domain knowledge to extract insights and support decision-making.

**Question 2**: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Answer:
*   **Overfitting** happens when a model learns the training data too well, including noise and irrelevant patterns. It performs well on training data but poorly on new/unseen data.
*   **Underfitting** occurs when a model is too simple and cannot capture the underlying patterns in the data. It performs poorly on both training and test sets.

    ***Detection:***
    1. Overfitting: High training accuracy, low testing accuracy
    2.  Underfitting: Low training accuracy, low testing accuracy

    ***Prevention Techniques:***
    1. Cross-validation (e.g., k-fold CV)
    2. Regularization techniques (L1/L2, dropout)
    3. Simplifying the model (for overfitting)
    4. Adding more data
    5. Early stopping in training
    6. Removing unnecessary features

    ***Bias-variance tradeoff plays a role:***
    1. Overfitting = low bias, high variance
    2. Underfitting = high bias, low variance

**Question 5:** Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

**Answer:**

Feature scaling ensures that all features contribute equally to distance-based and gradient-descent-based algorithms. Without scaling, large-value features dominate.


*   Min-Max Scaling (Normalization)

    Transforms values to 0–1 range.

    Good for: Neural Networks, KNN, distance-based models.
*   Standardization (Z-score scaling)

    Centers data around mean 0, std deviation 1.

    Good for: Linear Regression, SVM, Logistic Regression.

**Question 8:** Titanic Dataset

 a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?

 b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?

 **Answers :**

(a): Survival by Passenger Class

```python
import pandas as pd

# Load the titanic dataset, specifying the semicolon delimiter
titanic_df = pd.read_csv('titanic-passengers.csv', sep=';')

# Convert 'Survived' column from 'Yes'/'No' to 1/0 for numerical calculation
titanic_df['Survived'] = titanic_df['Survived'].map({'Yes': 1, 'No': 0})

# Calculate survival rates by Pclass
survival_by_pclass = titanic_df.groupby('Pclass')['Survived'].mean().reset_index()
survival_by_pclass['Survived'] = survival_by_pclass['Survived'] * 100 # Convert to percentage

print("Survival Rates by Passenger Class (Pclass):")
display(survival_by_pclass)

# Determine which class had the highest survival rate
highest_survival_class = survival_by_pclass.loc[survival_by_pclass['Survived'].idxmax()]
print(f"\nThe class with the highest survival rate is Pclass {int(highest_survival_class['Pclass'])} with a survival rate of {highest_survival_class['Survived']:.2f}%\n")

# Explanation for why this might have happened
print("**Reasoning for survival rates:**\n")
print("Generally, passengers in higher classes (e.g., Pclass 1) had better access to lifeboats, were located on higher decks, and were given priority during evacuation. This is often reflected in significantly higher survival rates for first-class passengers compared to second and third-class passengers. Socio-economic status, which correlates with Pclass, played a major role in the chances of survival during the Titanic disaster.")
```

(b): Age and Survival (Children vs Adults)

```python
titanic_df['Group'] = titanic_df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
age_survival = titanic_df.groupby('Group')['Survived'].mean()
print(age_survival)
```

Question 9: Flight Price Prediction Dataset

a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("flight_dataset.csv")

# Print columns to identify the correct price column (already confirmed 'Price')
print("DataFrame columns:", df.columns)

# Correct the price column name and convert to numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Construct Departure_Date from Date, Month, Year columns
df['Departure_Date'] = pd.to_datetime(df['Date'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Year'].astype(str), format='%d-%m-%Y')

# Define a hypothetical fixed booking date (e.g., beginning of the year for the flights)
# This helps in calculating 'days_left' consistently across all flights.
booking_date = pd.to_datetime('2019-01-01')

# Calculate days_left until departure (from the hypothetical booking date)
# Note: A negative 'days_left' would mean the booking date is after the departure date
# or the flight happened before our hypothetical booking date. We'll focus on positive days_left.
df['days_left'] = (df['Departure_Date'] - booking_date).dt.days

# Drop missing values in the relevant columns
df = df.dropna(subset=['days_left', 'Price'])

# Group by days_left and calculate the average price
avg_price = df.groupby('days_left')['Price'].mean().reset_index()

# Filter out any negative days_left if they resulted from the calculation
avg_price = avg_price[avg_price['days_left'] >= 0]

plt.figure(figsize=(12, 6))
plt.plot(avg_price['days_left'], avg_price['Price'], marker='o', linestyle='-')
plt.title("Average Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left Until Departure (from 2019-01-01 hypothetical booking date)")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

Question 10:  HR Analytics Dataset

a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).

b). Are employees with more projects more likely to leave?

**Answers:**

(a):Factors correlated with attrition

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("employee_attrition_test.csv")

# Assuming 'OverTime' acts as the attrition indicator and converting it to numeric
df['OverTime_Numeric'] = df['OverTime'].map({'Yes':1, 'No':0})

# Correlation with 'OverTime_Numeric'
cov = df.corr(numeric_only=True)['OverTime_Numeric'].sort_values(ascending=False)
print("Correlations with OverTime:")
print(cov)

# Plot satisfaction vs OverTime_Numeric
sns.boxplot(x='OverTime_Numeric', y='JobSatisfaction', data=df)
plt.title("OverTime vs Job Satisfaction")
plt.show()

# Plot OverTime (original categorical) to see its distribution
sns.countplot(x='OverTime', data=df)
plt.title("Distribution of OverTime")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("employee_attrition_test.csv")

# Use 'OverTime' as the attrition indicator and convert to numeric
df['Attrition_Indicator'] = df['OverTime'].map({'Yes':1, 'No':0})

# Group by 'JobLevel' as a proxy for projects and calculate the mean of the attrition indicator
proj_attrition = df.groupby('JobLevel')['Attrition_Indicator'].mean()

print(proj_attrition)
```
